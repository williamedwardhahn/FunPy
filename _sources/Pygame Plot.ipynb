# Pygame Plot

*Not for colab

```
import pygame
import numpy as np
from pygame import surfarray

pygame.init()

display = pygame.display.set_mode((256, 256))

while True:

    img = 255*np.random.random((256,256,3))

    surfarray.blit_array(display, img)
    pygame.display.flip()


pygame.quit()
```
