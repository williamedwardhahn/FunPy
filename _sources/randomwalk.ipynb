```
X = np.zeros((n,2)) 
for i in range(n - 1):
    X[i+1,:] = X[i,:] +  np.random.randn(1, 2)

plt.plot(X[:,0],X[:,1],'-.', markersize=15, marker='o')
```

```
X = np.zeros((n,2)) 
for i in range(n - 1):
    X[i+1,:] = X[i,:] +  np.random.randn(1, 2)


plt.scatter(X[:,0],X[:,1], marker='^', c=np.arange(n))
# plt.plot(X[:,0],X[:,1],'-.')
```

```
n = 10000
```
